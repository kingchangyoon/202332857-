```python
# 1. Google Drive에서 압축 파일 경로 설정
zip_path = '/content/drive/MyDrive/중간발표/서울특별시.Zip'  # 압축 파일 경로

# 2. 압축 해제
extracted_path = '/content/서울특별시'  # 압축 해제 폴더
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_path)

# 3. SHP 파일 경로 설정 (압축 해제된 폴더에서 SHP 파일 경로를 확인)
shp_path = os.path.join(extracted_path, '서울특별시.shp')  # 압축 해제된 SHP 파일 경로

# 4. SHP 파일 읽기
gdf = gpd.read_file(shp_path)

# 5. CSV 파일 읽기
csv_path = '/content/drive/MyDrive/중간발표/1인당 녹지면적5.csv'
df = pd.read_csv(csv_path)
```
